```python
import math
from PIL import Image
img = Image.open("image.png")
def rotate_image(img, angle):
    # Convert angle from degrees to radians
    radians_angle = math.radians(angle)

    # Get image dimensions
    width, height = img.size

    # Create a new image with the same dimensions as the original image
    rotated_img = Image.new('RGBA', (width, height), (255, 255, 255, 0))

    # Loop through each pixel in the new image
    for y in range(height):
        for x in range(width):
            # Calculate the corresponding pixel coordinates in the original image using trigonometry
            x_rotated = int((x * math.cos(radians_angle)) + (y * math.sin(radians_angle)))
            y_rotated = int((y * math.cos(radians_angle)) - (x * math.sin(radians_angle)))

            # Copy the color of the original pixel at the calculated coordinates
            if (x_rotated >= 0 and x_rotated < width and y_rotated >= 0 and y_rotated < height):
                color = img.getpixel((x_rotated, y_rotated))
                rotated_img.putpixel((x, y), color)

    return rotated_img
```

```python
import cv2
import numpy as np

# Load image
img = cv2.imread('vishu.jpg')

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Threshold the image to binary
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)

# Find contours
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Get the bounding box around each contour
boxes = [cv2.boundingRect(c) for c in contours]

# Initialize an empty list to hold the aligned characters
aligned_chars = []

# Loop through the boxes and align each character
for box in boxes:
    x, y, w, h = box
    
    # Crop the character from the image
    char = gray[y:y+h, x:x+w]
    
    # Get the centroid of the character
    M = cv2.moments(char)
    cX = int(M["m10"] / M["m00"])
    cY = int(M["m01"] / M["m00"])
    
    # Calculate the angle of rotation to align the character with the centroid
    angle = np.arctan2(cY-h//2, cX-w//2) * 180 / np.pi
    
    # Rotate the character
    rows, cols = char.shape
    M = cv2.getRotationMatrix2D((cols//2, rows//2), angle, 1)
    rotated = cv2.warpAffine(char, M, (cols, rows))
    
    # Add the aligned character to the list
    aligned_chars.append(rotated)

# Show the aligned characters
for char in aligned_chars:
    cv2.imshow('Aligned Character', char)
    cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = cv2.imread('image.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]

contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Create a new image with 3 dimensions to paste the rotated characters
result = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)

for cnt in contours:
    x, y, w, h = cv2.boundingRect(cnt)
    
    # Get the character from the binary image
    character = thresh[y:y+h, x:x+w]
    
    moments = cv2.moments(character)
    if moments["m00"] != 0:
        cx = moments["m10"] / moments["m00"]
        cy = moments["m01"] / moments["m00"]
        theta = np.degrees(np.arctan2(cy - h/2, cx - w/2))
    else:
        theta = 0
    
    rot_mat = cv2.getRotationMatrix2D((w/2, h/2), theta, 1.0)
    rotated = cv2.warpAffine(character, rot_mat, (w, h), flags=cv2.INTER_CUBIC, borderMode=cv2.BORDER_REPLICATE)
    
    # Create a new image to paste the rotated character
    char_img = np.zeros((h, w, 3), dtype=np.uint8)
    char_img[:, :, 0] = rotated
    char_img[:, :, 1] = rotated
    char_img[:, :, 2] = rotated
    
    # Rotate the character to a fixed angle (20 degrees in this example)
    rot_mat = cv2.getRotationMatrix2D((w/2, h/2), 20, 1.0)
    rotated = cv2.warpAffine(char_img, rot_mat, (w, h), flags=cv2.INTER_CUBIC, borderMode=cv2.BORDER_REPLICATE)
    
    # Calculate the offset to paste the rotated character into the new image
    offset_x = x + w // 2 - rotated.shape[1] // 2
    offset_y = y + h // 2 - rotated.shape[0] // 2
    
    result[offset_y:offset_y+rotated.shape[0], offset_x:offset_x+rotated.shape[1]] = rotated

cv2.imshow('Result', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
